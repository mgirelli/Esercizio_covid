```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('data_covid.csv', sep=',', index_col='data', parse_dates=True)
```

```python
dataframe.head()
```

```python
class plot_covid:
    def __init__(self, data):
        self.dataframe = data
    def plot_una_regione(self, nome, value):
        self.regione = nome
        self.ricovero = value
        data_tagliato = self.dataframe.loc[self.dataframe.denominazione_regione == self.regione]
        plt.plot(data_tagliato.index, data_tagliato[self.ricovero], linewidth=1.0)
        plt.show()
    def plot_ogni_regione(self, value):
        for nome, dati in self.dataframe.groupby('denominazione_regione'):
            plt.plot(dati.index, dati[value], linewidth=1.0, label = f'{nome}')
        plt.legend(prop={'size': 6})
        plt.show()
        
        
```

```python
obj = plot_covid(dataframe)
```

```python
obj.plot_una_regione('Veneto', 'ricoverati_con_sintomi')
```

```python
obj.plot_ogni_regione('ricoverati_con_sintomi')
```
